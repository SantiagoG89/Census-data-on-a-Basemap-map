```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch

#¿donde vive el productor CREA?
fig = plt.figure(figsize=(22,12))
ax      = fig.add_subplot(111)
base = pd.read_excel("C:/Users/user/Documents/DATOS PARA MAPA.xlsx")


m = Basemap(projection='merc',
           llcrnrlat = -47,
           urcrnrlat = -21,
           llcrnrlon = -73,
           urcrnrlon = -53,
           resolution = 'f')

m.drawcoastlines()
m.drawcountries()

    

m.scatter((base["LON"]).tolist(), (base["LAT"]).tolist(),c="green", s = ((base["Size"]*5).tolist()), latlon=True)




m.readshapefile("C:/Users/user/Documents/Mapa_Solo_RegionCREA_wgs84","Regiones",  linewidth=1)



patches = []

for shape in m.Regiones:
        patches.append(Polygon(np.array(shape), True) )
        
        


colors = ["#999999", "#000000"]
        
ax.add_collection(PatchCollection(patches, facecolor= colors, edgecolor="k", linewidths=1., zorder=2, alpha = 0.2))

plt.savefig('C:/Users/user/Pictures/mapa1.png')














```
